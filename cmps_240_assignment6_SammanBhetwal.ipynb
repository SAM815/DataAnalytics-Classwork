```python
#Question number 2
import matplotlib.pyplot as plt

homeownership = loans_df['homeownership'].value_counts()
homeownership.plot(kind='bar')
plt.show()
```

```python
#Question 3
home_grade_count = pd.crosstab(loans_df['homeownership'], loans_df['grade'])
home_grade_count.plot(kind='bar', stacked=True)
plt.show()
```

```python
#Question number 4
home_grade_grouped = loans_df.groupby(['homeownership', 'grade']).size().unstack()
home_grade_grouped.plot(kind='bar')
plt.show()
```

```python
#Question number 5
```

The stacked bar plot in question 3 is more appropriate because it shows the distribution of the â€˜gradeâ€™ variable within each level of the â€˜homeownershipâ€™ variable. However, the grouped bar plot in question 4 does not convey the same information as well because it separates the bars by grade rather than grouping them by homeownership.

```python
#Question number 6
home_grade_count = pd.crosstab(loans_df['homeownership'], loans_df['grade'])
home_grade_count.plot(kind='bar', stacked=True)
plt.xlabel('Homeownership')
plt.ylabel('Frequency')
plt.title('Distribution of Loan Grade by Homeownership')
plt.show()
```
