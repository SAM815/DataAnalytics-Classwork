### imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Question number 1

```python
loans_df = pd.read_csv("loans.csv")
loans_df
```

### Question number 2

```python
plt.hist(loans_df['loan_amount'])
plt.xlabel('Loan Amount ($)')
plt.ylabel('Count')
plt.title('Histogram of Loan Amounts')
plt.show()
```

### Question number 3

```python
plt.hist(loans_df['loan_amount'], bins=range(0, 40000, 1000))
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Loan Amount')
plt.show()
```

```python
plt.hist(loans_df['loan_amount'], bins=range(0, 40000, 5000))
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Loan Amount')
plt.show()
```

```python
plt.hist(loans_df['loan_amount'], bins=range(0, 40000, 20000))
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Loan Amount')
plt.show()
```

Binwidth of 5000 is the most appropriate as it captures the overall shape of the distribution while still providing enough detail to see the peak and tails of the distribution. A binwidth of 1000 may be too fine-grained and may result in too many small bins, while a binwidth of 20000 may be too broad and may miss some of the finer details in the distribution.

### Question number 4

```python
plt.hist(loans_df[loans_df['homeownership'] == 'MORTGAGE']['loan_amount'], bins=range(0, 40000, 5000), label='MORTGAGE')
plt.hist(loans_df[loans_df['homeownership'] == 'RENT']['loan_amount'], bins=range(0, 40000, 5000), label='RENT')
plt.hist(loans_df[loans_df['homeownership'] == 'OWN']['loan_amount'], bins=range(0, 40000, 5000), label='OWN')

plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Loan Amount by Homeownership')
plt.legend(loc='upper right')

plt.show()
```

### Question number 7

```python

interest_rate = loans_df['interest_rate']

# create the boxplot using Seaborn
sns.boxplot(x=interest_rate)
plt.title('Interest Rate Boxplot')
plt.xlabel('Interest Rate')
plt.show()
```

So, in the box plot we can see the some values outside the maximum values. These values are called
outliers as they lie outside the typical range of values.


### Question number 8

```python

sns.boxplot(x='grade', y='interest_rate', 
data=loans_df,order=sorted(loans_df['grade'].unique()))
plt.title('Interest Rate Boxplot by Loan Grade')
plt.xlabel('Loan Grade')
plt.ylabel('Interest Rate')
plt.show()
```

The boxplot above shows the interest_rate variable as it varies with the variable grade. We can
see the low interest rate in Grade A and how the interest rate increases as the Loan grade increases signalling their relationship.

### Question number 9

```python

annual_income = loans_df.iloc[:, 6]
sns.boxplot(x='homeownership', y='annual_income', 
data=loans_df,order=sorted(loans_df['homeownership'].unique()))
plt.title('Annual Income Boxplot by Homeownership')
plt.xlabel('Homeownership')
plt.ylabel('Annual Income')
plt.show()
```

In the box plot above we can compare the annual income of people in relation to their Homeownership. We can see that most people with high annual income have mortage which is followed by
rent and finally the people with less income own a property.
